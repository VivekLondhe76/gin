```python
# 1. Import Libraries

# pandas for data manipulation and analysis.
import pandas as pd

# Imports NumPy for numerical operations.
import numpy as np

#Imports Matplotlib for creating plots.
import matplotlib.pyplot as plt

#Imports Seaborn for advanced visualizations
import seaborn as sns

# LabelEncoder: Encodes labels (e.g., 'Male', 'Female') into numbers.
# StandardScaler: Normalizes numerical data.
from sklearn.preprocessing import LabelEncoder, StandardScaler
%matplotlib inline
sns.set(style="whitegrid")

df = pd.read_csv('C:/Users/ASUS/Desktop/ML_K/tested.csv')

```

```python
#  2. Load the Dataset
#  Displays the first 5 rows
df.head() 
```

```python
#  3. Overview of Data

# Displays rows and column
print("Shape : ", df.shape)

# Lists all column names
print("\nColumns:\n\n", df.columns)

# Provides a summaryâ€”data types, non-null counts, memory usage
print("\nInfo:\n" )
df.info()

# Shows statistics for both numeric and non-numeric data
print("\nSummary Statistics:\n")
df.describe(include='all')
```

```python
#  4. Checking Missing Data


#Counts missing values per column.
#Returns (index,value) = (column,no_of_missing_values)
missing = df.isnull().sum() 

# The print shows only columns with missing values.
print("\nMissing Values:\n", missing[missing > 0])

# Visualizing missing data
# Yellow = Missing , Purple = Not Missing
plt.figure(figsize=(10,6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title("Missing Data Heatmap")
plt.show()
```

```python
#  5. Handling Missing Data

# Example: Fill numeric with median, categorical with mode
for column in df.columns:
    if df[column].dtype == 'object':
        df[column].fillna(df[column].mode()[0], inplace=True)
    else:
        df[column].fillna(df[column].median(), inplace=True)

# Check again
print(df.isnull().sum())
```

```python
#  6. Visualization

# Countplot for a categorical column (edit based on actual columns)
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Sex')  # Replace 'Sex' with actual categorical column
plt.title("Count of Categories")

# Distribution of numerical column
plt.figure(figsize=(6,4))
sns.histplot(df['Age'], kde=True)  # Replace 'Age' with numeric column
plt.title("Distribution of Age")

# Boxplot by category
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x='Sex', y='Age')  # Replace as needed
plt.title("Boxplot: Age by Sex")
```

```python
#  7. Correlation & Heatmap

plt.figure(figsize=(10,6))

# df.corr(numeric_only=True): Calculates correlation between numeric columns.
corr_matrix = df.corr(numeric_only=True) 

# sns.heatmap: Visualizes this correlation matrix.
# annot=True: Shows numeric values inside cells.
# coolwarm: Red/blue color scheme indicating correlation strength.
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5) 
plt.title("Correlation Heatmap")
plt.show()
```

```python
#  Scatter Plot with Hue (e.g., based on Survival or Gender)
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='Age', y='Fare', hue='Survived')  # Replace 'Survived' with your categorical column
plt.title('Scatter Plot: Age vs Fare (Colored by Survived)')
plt.show()
```

```python
#  8. Encoding Categorical Variables

# Label Encoding for binary columns
# "male" â†’ 1, "female" â†’ 0
label_enc = LabelEncoder()
for column in df.select_dtypes(include='object').columns:
    if df[column].nunique() == 2:
        df[column] = label_enc.fit_transform(df[column])

# One-Hot Encoding for others
df = pd.get_dummies(df, drop_first=True)

df.head()
```

```python
# 9. Feature Scaling

scaler = StandardScaler()
num_cols = df.select_dtypes(include=np.number).columns
df[num_cols] = scaler.fit_transform(df[num_cols])

df.head()
```
